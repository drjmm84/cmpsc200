```python
# import math
import matplotlib.pyplot as plt

def collatz(x):
    n: int
    mx: int
    mx = x
    n = 0
    while x > 1:
        if x % 2 == 0:
            x = x / 2
        else:
            x = 3 * x + 1
        if x > mx:
            mx = x
        n = n + 1
    return mx, n


Z = collatz(5)
print(Z)
```

```python
Nlist = []
mxlist = []
for k in range(1,10000+1):
    Z = collatz(k)
    Nlist.append(Z[0])
    mxlist.append(Z[1])
#plt.loglog(Nlist, mxlist,'.')
plt.loglog(Nlist,'.')
plt.show()
```
