```python
import matplotlib.pyplot as plt
import numpy

q = numpy.array([])
samps = int(5e4)
for k in range(0, samps):
    X = 1.2 + 0.18 * numpy.random.randn(260 * 15)
    Y = 1.1 + 0.4 * numpy.random.random(140 * 15)
    q = numpy.append(q, numpy.sum(X) + numpy.sum(Y))

m = numpy.cumsum(q) / numpy.arange(1, samps + 1)
plt.semilogx(numpy.arange(1, samps + 1), m)
plt.xlabel("N")
plt.ylabel("Average Weight [lbs]")
plt.grid()
plt.show()

print(m[samps - 1])

overload = sum(q > 7430.0) / samps
print(overload)
```
