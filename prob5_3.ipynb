```python
import matplotlib.pyplot as plt
import numpy

q = []
samps = int(3e5)
for k in range(0, samps):
    A = numpy.random.random([12, 1])
    B = numpy.random.random([7, 1])
    if numpy.sum(A < 0.98) == 12 and numpy.sum(B < 0.91) >= 5:
        q.append(1)
    else:
        q.append(0)

m = numpy.cumsum(q) / numpy.arange(1, samps + 1)
plt.semilogx(numpy.arange(1, samps + 1), m)
plt.xlabel("N")
plt.ylabel("Average")
plt.grid()
plt.show()
```
